### Replacement Players

Finding replacement players:

Instatiating (representing/creating an instance of) a dictionary with positions as our KEYS, and empty strings as VALUES then iterate over dataframe. Then update empty string values in our dictionary, based off of our data frame of replacement players. Then look into our projection_df and find fantasy points scored. Fantasy points scored are replacement values for the positions.

```python
replacement_players = {
    'RB':'',
    'QB':'',
    'WR':'',
    'TE':''
}
replacement_players
```

iterrows turns df into an iterable, row by row and allow us to pull out values
like .items allowed us to iterate over a dictionary

Typical format is "for index in row" or "for idx, row in" - this is unpacking notation.
Similar to key, value for dict...but this is now index,row for df.
_ is good pythonic convention for a placeholder value.
Could use "for idx, row in" but this is bad style/convention.


If Pos (now known as adp_df_position) in adp_df_cutoff exists as one of the keys in the replacement_players dict, then replace the Pos (adp_df_position) with Player (adp_df_player) from adp_df_cutoff within the replacement_players dict.

adp_df_position is being used as an indexer in the replacement_players dict, meaning we're now using adp_df_position to match to the keys of the dict.
The keys of the dict are currently unnamed/unlabelled.

If this specific adp_df_position exists as a 'key' in the dict, then the 'value' for that 'key' will be replaced with the adp_df_player in the same row
as adp_df_position in adp_df_cutoff.

This loop keeps running through all rows until it reaches the last row of the position type. It will then pull that player. Then it moves onto next position type and repeats the process.

```python
for index,row in adp_df_cutoff.iterrows():
  #extract specific Pos in a row, and specific Player of a row, and name each
  adp_df_position = row['Pos']
  adp_df_player = row['Player']

  if adp_df_position in replacement_players:
    replacement_players[adp_df_position] = adp_df_player
```

### Value over Replacement (VOR)

```python
df_vor.head()
```

####apply()
Apply takes in callable (aka a function) as first argument to map/apply across the entire df

Second arg is axis (a keyword arg)

####the lambda function
The arguement is row. We then get the FantasyPoints for each row of the df.  We also get the Pos of each row in the df. We use the Pos from the df row as the indexer for the replacement_values dict.

So for Josh Allen - we grab FantasyPoints of 413.44 and POS of QB. QB is then input as the key we're searching for in replacement_values dict. If QB is present in the dict, then pull the value. For QB that is 295.36.

So we have VOR = 413.44-295.36

axis=1 allows us to apply this across all rows of df, whereas axis=0 applies function across all df columns

so using apply() with axis=1, we are able to apply this lambda function across all rows of df

```python
df_vor['VOR']=df_vor.apply(
    lambda x: x['FantasyPoints'] - replacement_values[x['Pos']], axis=1
)

#could also do lambda row: row['FantasyPoints'] - replacement_values[row['Pos']], axis=1
df_vor.head()
```

```python
#sort df by VOR. Have to reassign df to save sorting changes!
df_vor = df_vor.sort_values(by='VOR', ascending=False)

df_vor.head()
```

Create VOR rank column, display first 100.  Need to alter an option in pandas display settings to actually see all 100. Without altering, we only see first and last 5 of 100.

```python
df_vor['VOR Rank'] =  df_vor['VOR'].rank(ascending=False)
pd.set_option('display.max_rows', None)

df_vor.head()
```

Group by position, then grab the VOR and show descriptive stats for each position's VOR.

We see that RBs are most valuable, with lowest mean VOR. This means that the RB position will add the most (or have the least disparity in) VOR. Meaning it's easier to find a replacement among RBs

```python
df_vor.groupby('Pos')['VOR'].describe()
```

####VOR normalizing
Now we can normalize to scale the data between 0 to 1, to better visualize.
Using the formula VORnorm = VORmin / (VORmax-VORmin).

Aka VORnorm = VORmin / (range of VOR values)

```python
df_vor['VORnorm'] = df_vor['VOR'].apply(lambda x: (x - df_vor['VOR'].min()) / (df_vor['VOR'].max() - df_vor['VOR'].min()))
df_vor.head()
```

####VOR viz
Box shows the quartiles. Whiskers show the rest of distribution with the single points being outliers. Outliers determined by function of the inter-quartile range (>1.5xIQR above the third quartile, or <1.5xIGR below the first quartile)

```python
sns.boxplot(x=df_vor['Pos'], y=df_vor['VORnorm'])
```

##Join df_vor & adp_df for final_df

Then want to find players that are undervalued relative to ADP.

```python
df_vor.head()
```

Ditch VOR column since we just want to use VORnorm from here on out

```python
df_vor = df_vor[['Player','Pos','Team','FantasyPoints','VORnorm','VOR Rank']]
df_vor.head()
```

Rename columns in df_vor using rename() which takes in a "mapper" which is essentially a dictionary. So we can enter our column names into this dictionary, and enter new column names as the values while old column names are keys.

```python
df_vor = df_vor.rename({
    'VORnorm': 'Value',
    'VOR Rank': 'Value Rank'
}, axis = 1)

df_vor.head()
```

Rename columns in adp_df.

Replace values in the Player column of adp_df.

Delete 'Team' from adp_df since it's a duplicate column. axis=1 drops the label from the column rather than the index (row)

We do this to make sure all the data in the two data frames matches up properly.

```python
adp_df = adp_df.rename({
    'PLAYER': 'Player',
    'POS': 'Pos',
    'AVG': 'Average ADP',
    'ADP RANK': 'ADP Rank'
}, axis=1)

adp_df['Player'] = adp_df['Player'].replace({
    'Kenneth Walker III': 'Kenneth Walker',
    'Travis Etienne Jr.': 'Travis Etienne',
    'Brian Robinson Jr.': 'Brian Robinson',
    'Pierre Strong Jr.': 'Pierre Strong',
    'Michael Pittman Jr.': 'Michael Pittman',
    'A.J. Dillon': 'AJ Dillon',
    'D.J. Moore': 'DJ Moore'
})

adp_df = adp_df.drop('Team',axis=1)

adp_df.head()
```

Merge adp_df with df_vor to make final_df

```python
final_df = df_vor.merge(adp_df, how='left', on=['Player','Pos'])
final_df.head()
```

Now calculate difference bewteen ADP Rank and Value Rank.

```python
final_df['ADP_Val_diff'] = final_df['ADP Rank']-final_df['Value Rank']
final_df.head()
```

Sort to find top sleepers = highest (most positive) difference between ADP and Value Rank.

If the ADP Rank is very high, then they are drafted late. If their Value Rank is very low then they have great value. So high ADP Rank - low Value Rank results in a player who is undervalued (drafted late even though great value)

```python
final_df.sort_values(by='ADP_Val_diff', ascending=False)[:10]
```

Most overvalued = lowest (most negative) difference. These are players ddrafted earlier than their Value Rank merits.

```python
final_df.sort_values(by='ADP_Val_diff', ascending=True)[:10]
```
